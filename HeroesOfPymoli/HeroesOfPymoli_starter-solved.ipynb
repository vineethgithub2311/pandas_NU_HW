
## Purchasing Analysis (Gender)

* Create a summary data frame to hold the results

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_df = purchase_data
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 1000]
group_labels = ["<10", "10-14", "15-19","20-24","25-29","30-34","35-39","40+"]
purchase_df['Range'] = pd.cut(purchase_data['Age'], bins, labels=group_labels)

purchase_count = purchase_df.groupby(['Range']).sum()["Price"].rename("Total Purchase Value")
avg_purchase_price = purchase_df.groupby(['Range']).mean()["Price"].rename("Average Purchase Value")
purchase_age_count = purchase_df.groupby(['Range']).count()["Price"].rename("Purchase Age Count")
avg_purchase_person = purchase_df.groupby(['Range']).mean()["Price"].rename("Avg Total Purchase per Person")

#dataframe
purchase_df2 = pd.DataFrame({"Purchase Count": purchase_age_count, 
                             "Average Purchase Price": avg_purchase_price, 
                             "Total Purchase Value": purchase_count, 
                             "Avg Total Purchase per Person":avg_purchase_person })
#formatting
purchase_df2["Total Purchase Value"] = purchase_df2["Total Purchase Value"].map("${:,.2f}".format)
purchase_df2["Average Purchase Price"] = purchase_df2["Average Purchase Price"].map("${:,.2f}".format)
purchase_df2 ["Purchase Count"] = purchase_df2["Purchase Count"].map("{:,}".format)
purchase_df2["Avg Total Purchase per Person"] = purchase_df2["Avg Total Purchase per Person"].map("${:,.2f}".format)
purchase_df2 = purchase_df2.loc[:, ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]

purchase_df2
```
